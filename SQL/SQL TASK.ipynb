# 4. dump all the data into car table

```python
file1 = open('car.data','r')
```

```python
file1.read()
```

```python
import csv

with open('car.data','r') as f:
    car_data = csv.reader(f,delimiter='\n')
    for i in car_data:
        row_data = i[0].split(',')
        query = """insert into cardataset.car(buying,maint,doors,persons,lugboot,price,safery) values(%s, %s, %s, %s, %s, %s, %s)"""
        cursor.execute(query,row_data)
mydb.commit()
```
